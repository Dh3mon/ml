## Tipe Data Python

Pada bagian ini ditampilkan beberapa tipe data dasar dalam Python yang sering digunakan dalam Machine Learning seperti number, string, list, tuple, dictionary, dan boolean.

```python
# ===== TIPE DATA UNTUK MACHINE LEARNING =====

# 1. Numbers (int, float)
age = 20
height = 180.2

print(f"Age: {age}, Height: {height}")
```

### Penjelasan

Kode di atas membuat dua variabel yaitu `age` dan `height`.

- `age` merupakan tipe data **integer (int)** yang menyimpan nilai umur.
- `height` merupakan tipe data **float** yang menyimpan angka desimal.

Fungsi `print()` digunakan untuk menampilkan nilai variabel tersebut ke layar.

```python
# 2. Strings
name = "Machine Learning"

print(f"Course: {name}")
```

### Penjelasan

Kode ini menggunakan tipe data **string**.

Variabel `name` menyimpan teks "Machine Learning".  
Tipe data string digunakan untuk menyimpan data berupa teks atau label.

```python
# 3. Lists (ordered, mutable)
scores = [85, 90, 78, 92]

print(f"Scores: {scores}")
print(f"First score: {scores[0]}")
```

### Penjelasan

Kode ini membuat **list** bernama `scores`.

List adalah tipe data yang digunakan untuk menyimpan beberapa nilai dalam satu variabel.

- `[85, 90, 78, 92]` merupakan isi dari list
- `scores[0]` digunakan untuk mengambil nilai pertama dari list

```python
# 4. Tuples (ordered, immutable)
coordinates = (40.7128, 74.0060)

print(f"Coordinates: {coordinates}")
```

### Penjelasan

Kode ini menggunakan tipe data **tuple**.

Tuple hampir sama seperti list, tetapi **tidak bisa diubah (immutable)** setelah dibuat.

Pada contoh ini tuple digunakan untuk menyimpan koordinat latitude dan longitude.

```python
# 5. Dictionaries (key-value pairs)

student = {
    'name': 'Rafly',
    'age': 20,
    'ipk': 3.8
}

print(f"Student: {student['name']}, IPK: {student['ipk']}")
```

### Penjelasan

Dictionary digunakan untuk menyimpan data dalam bentuk **key dan value**.

Pada kode di atas:

- `name`, `age`, dan `ipk` adalah **key**
- Nilainya adalah **value**

Data di dalam dictionary dapat diakses menggunakan nama key, seperti `student['name']`.

```python
# 6. Boolean

is_student = True
has_experience = False

print(f"Student: {is_student}, Experience: {has_experience}")
```

### Penjelasan

Boolean adalah tipe data yang hanya memiliki dua nilai yaitu:

- `True`
- `False`

Biasanya digunakan dalam kondisi logika atau percabangan program.

## Loop dan Control Flow

Loop digunakan untuk menjalankan suatu perintah secara berulang pada sekumpulan data.

```python
# ===== FOR LOOP =====

scores = [85, 90, 78, 92]

for score in scores:
    print(f"Score: {score}")
```

### Penjelasan

Kode di atas menggunakan **for loop** untuk menampilkan setiap nilai dalam list `scores`.

Perulangan akan berjalan satu per satu sampai semua elemen dalam list selesai diproses.

```python
# Loop dengan index

for i, score in enumerate(scores):
    print(f"Student {i+1}: {score}")
```

### Penjelasan

Fungsi `enumerate()` digunakan untuk mendapatkan **index dan nilai sekaligus** dari sebuah list.

- `i` adalah index
- `score` adalah nilai dari list

```python
# WHILE LOOP

count = 0

while count < 3:
    print(f"Count: {count}")
    count += 1
```

### Penjelasan

Perulangan **while** akan terus berjalan selama kondisi bernilai benar.

Pada contoh ini, loop akan berhenti ketika nilai `count` sudah mencapai 3.

```python
# IF ELSE

for score in scores:
    if score >= 85:
        print(f"Score {score}: EXCELLENT")
    elif score >= 75:
        print(f"Score {score}: GOOD")
    else:
        print(f"Score {score}: NEEDS IMPROVEMENT")
```

### Penjelasan

Kode ini menggunakan **percabangan if, elif, dan else**.

Program akan mengecek nilai `score` dan menentukan kategori nilai berdasarkan kondisi yang diberikan.

## Function

Function digunakan untuk membuat blok kode yang dapat digunakan kembali.

```python
def greet(name):
    return f"Hello, {name}!"

print(greet("Rafly"))
```

### Penjelasan

Function `greet()` dibuat untuk memberikan salam kepada seseorang.

Parameter `name` digunakan sebagai input, kemudian function akan mengembalikan teks salam.

```python
def calculate_average(scores):
    total = sum(scores)
    count = len(scores)
    return total / count

scores = [85, 90, 78, 92]

avg = calculate_average(scores)

print(f"Average score: {avg}")
```

### Penjelasan

Function `calculate_average()` digunakan untuk menghitung rata-rata nilai.

- `sum()` digunakan untuk menjumlahkan semua nilai
- `len()` digunakan untuk menghitung jumlah data

```python
square = lambda x: x ** 2

print(square(3))
```

### Penjelasan

Lambda function adalah function sederhana yang ditulis dalam satu baris.

Pada contoh ini function digunakan untuk menghitung **kuadrat dari sebuah angka**.
