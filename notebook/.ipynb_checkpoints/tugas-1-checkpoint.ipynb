### Tipe Data Number

```python
bb = 60
tb = 190
print(f"Berat Badan : {bb}, Tinggi Badan {tb}")
```

### Strings

```python
nama = "Fajar"
jurusan = "Informatika"

print(f"Nama : {nama}, Jurusan : {jurusan}")
```

### Boolean

```python
lulus = True

print(f"Status Kelulusan : {lulus}")
```

### List

nilai = [80, 75, 90, 85]

print(f"Daftar Nilai : {nilai}")
print("Nilai pertama :", nilai[0])

### Dictionary

```python
mahasiswa = {
    "nama": "Fajar",
    "umur": 20,
    "ipk": 3.8
}
print(f"Data Mahasiswa : {mahasiswa}")
print("Nama Mahasiswa :", mahasiswa["nama"])
```

### Percabangan

```python
nilai = 80

if nilai >= 75:
    print(f"Nilai {nilai} : Lulus")
else:
    print(f"Nilai {nilai} : Tidak Lulus")
```

### Perulangan

```python
for i in range(5):
    print(f"Perulangan ke-{i}")
```

### Perulangan Pada Array

```python
nilai_list = [80, 65, 90, 70]

for nilai in nilai_list:
    print(f"Nilai : {nilai}")
```
