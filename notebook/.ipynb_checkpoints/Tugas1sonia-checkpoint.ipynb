**Tipe Data**

#### 1. Numbers (int, float) - untuk nilai numerik seperti umur atau harga

```python
umur = 10
tinggi = 157
print(f"Umur: {umur}, Tinggi: {tinggi}")
```

#### 2. Strings - Digunakan untuk label, kategori, atau nama

```python
nama = "Sonia"
print(f"Nama: {nama}")
```

#### 3. Lists (ordered, mutable) - Menyimpan urutan data yang bisa diubah

```python
nilai = [85, 90, 78, 92]
print(f"Nilai: {nilai}")
print(f"Nilai pertama: {nilai[0]}")
```

#### 4. Tuples (ordered, immutable) - Menyimpan data koordinat atau konstanta yang tidak boleh berubah

```python
koordinat = (40.7128, 74.0060) # (latitude, longitude)
print(f"Koordinat: {koordinat}")
```

#### 5. Dictionaries (key-value pairs) - Menyimpan data terstruktur seperti profil objek

```python
murid = {
    'nama': 'sonia',
    'umur': 10,
    'ipk': 4.0
}
print(f"Mahasiswa: {murid['nama']}, IPK: {murid['ipk']}")
```

#### 6. Boolean - Digunakan untuk logika pengkondisian (True/False)

```python
is_mhs = True
has_experience = False
print(f"Mahasiswa: {is_mhs}, Experience: {has_experience}")
```

**Loop dan Control Flow**

#### 1. FOR LOOP - untuk iterasi pada list atau array data science

```python
nilai = [85, 90, 78, 92]
for score in nilai:
    print(f"Nilai: {score}")
```

#### 2. Loop dengan index menggunakan enumerate

```python
for i, nilai in enumerate(nilai):
    print(f"Mahasiswa {i+1}: {nilai}")
```

#### 3. WHILE LOOP - Berjalan selama kondisi terpenuhi

```python
print("\n=== WHILE LOOP ===")
hitung = 0
while hitung < 3:
    print(f"Hitung: {hitung}")
    hitung += 1
```

#### 4. CONTROL FLOW (if/else) - Digunakan dalam tahap preprocessing untuk klasifikasi data sederhana

```python
# Definisi variabel scores harus ada sebelum digunakan
scores = [85, 90, 78, 92] 

print("\n=== IF/ELSE ===")
for score in scores:
    if score >= 85:
        print(f"Nilai {score}: Sangat Baik")
    elif score >= 75:
        print(f"Nilai {score}: Baik")
    else:
        print(f"Nilai {score}: Tingkatkan Lagi")
```

**Function**

#### Basic function dengan return value

```python
def greet(nama):
    """menyapa seseorang"""
    return f"Hello, {nama}!"

print(greet("Sonia"))
```

#### Function dengan parameter list untuk menghitung rata-rata

```python
def rata2(scores):
    total = sum(scores)
    count = len(scores)
    return total / count

avg = rata2([85, 90, 78, 92])
print(f"Nilai Rata-rata: {avg}")
```

#### Lambda function untuk operasi matematika cepat dan anonim

```python
kotak = lambda x: x ** 2
print(f"squared(3) adalah: {kotak(3)}")
```
